```python
import pandas as pd

df = pd.read_csv("./ais_with_time.csv")

df = df[df['speed'] <= 30]
df.fillna(0)
# print(df)
```

```python
first = df.iloc[0]['vessel_name']

times = []
for i in range(0, len(df)):
    if df.iloc[i]['vessel_name'] != first or i == 0:
        first = df.iloc[i]['vessel_name']
        times.append(0)
        continue
    final = df.iloc[i]['timestamp'][11:19]
    initial = df.iloc[i-1]['timestamp'][11:19]
    # print(final, initial)
    final = int(final[0:2]) * 3600 + int(final[3:5]) * 60 + int(final[6:8])
    initial = int(initial[0:2]) * 3600 + int(initial[3:5]) * 60 + int(initial[6:8])
    timediff = final - initial
    # print(timediff)
    if timediff > 3*3600:
        timediff = 3*3600
    times.append(float(timediff))


df['time_difference'] = times
df
# df.to_csv("test.csv")
    
```

```python
# in Transit: If speed > 1 OR navigational status = 0, 3, 4, 8, 11, 12 (considered 'moving' statuses).
df = df[(df["speed"] > 1) | (df["nav_stat"] == 0) | (df["nav_stat"] == 3) | (df["nav_stat"] == 4) | (df["nav_stat"] == 8) | (df["nav_stat"] == 11) | (df["nav_stat"] == 12)]

ship = df.iloc[0]['vessel_name']
totaltime = 0
noOfShips = 1

for i in range(len(df)):
    if df.iloc[i]['vessel_name'] != ship:
        noOfShips += 1
        ship = df.iloc[i]['vessel_name']
        continue
    totaltime += df.iloc[i]['time_difference']
    
print(totaltime/noOfShips/3600)
```

```python
df['max_speed'] = 1.066 * df['vref']

df['me_ef'] = df['sfc_me'] * 0.867 * 3.667
df['ae_ef'] = df['sfc_ae'] * 0.867 * 3.667
df['ab_ef'] = df['sfc_ab'] * 0.867 * 3.667
```

```python
# E = LA X A x EF

# Where
# E = Emissions (grams [g])
# La = Auxiliary engine load (kW)
# A = Activity (hours [h]) (hours/call)
# EF = Emission Factor (grams per kilowatt-hour [g/kWh])

df['e_ae'] = df['ael'] * (df['time_difference']/3600) * df['ae_ef']

# E = LB X A x EF

# Where
# E = Emissions (grams [g])
# L
# B
# L 
# B
# ​
#   = Boiler Load (kW),
# A = Activity (hours [h]) (hours/call)
# EF = Emission Factor (grams per kilowatt-hour [g/kWh])

df['e_ab'] = df['abl'] * (df['time_difference']/3600) * df['ab_ef']

df.to_csv("test.csv")
```

```python
dist = pd.read_csv("./dataset/distance_dataset.csv")
dist = dist.drop_duplicates(subset='imo', keep='last')

# Assume df is already defined and loaded
print(len(df))
# Merge DataFrames on 'imo' column
df = df.merge(dist[['imo', 'avg_speed_kt', 'max_speed_kt']], on='imo', how='left', suffixes=('', '_new'))

# df.to_csv("test.csv")
```

```python
# LF=(AS/MS)^3
# Where
# LF = Load Factor
# AS = Average speed (knots)
# MS = Maximum speed (knots)

# Calculate Load Factor (LF)
df['lf'] = (df['avg_speed_kt'] / df['max_speed_kt'])**3 * 10
df['lf'] = df['lf'].fillna(0)
print(df.isnull().sum())

# Load the supplementary table
llaf = pd.read_csv("./supplementary_tables/llaf_table.csv")

# Ensure 'llaf' column exists in df
df['llaf'] = 0  # Initialize with default values

# Iterate and assign 'llaf' based on conditions
for i in range(len(df)):
    lf_value = int(df.iloc[i]['lf'])
    if lf_value >= 19 or lf_value < 2:
        df.at[i, 'llaf'] = 1
    else:
        # Ensure the index is within the bounds of the 'llaf' DataFrame
        if 2 <= lf_value <= len(llaf) + 1:
            df.at[i, 'llaf'] = llaf.iloc[lf_value - 2]['CO2']

# Calculate 'e_me'
df['e_me'] = df['p'] * df['lf'] / 100 * df['me_ef'] * df['llaf']

df.to_csv("traversal_JIT_2.csv")
```

```python
# Total emissions before JIT
df[['e_ae', 'e_ab', 'e_me']] = df[['e_ae', 'e_ab', 'e_me']].fillna(0)

total = 0

for i in range(len(df)):
    total += df.iloc[i]['e_ae'] + df.iloc[i]['e_ab'] + df.iloc[i]['e_me']

print(total)
```
